```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold.t_sne import TSNE
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
import joblib as jb
```

```python
best_U = jb.load( 'chicago_bias_GP_decay_best_U_rank_2.jb')
#best_U = jb.load('slc_bias_GP_decay_best_U_rank_2.jb')
#best_U = jb.load('911_bias_GP_decay_best_U_rank_2.jb')
#best_U = jb.load( 'article_bias_GP_decay_best_U_rank_2.jb')[1:]
```

```python
beats = best_U[0]
types = best_U[1]
```

```python
beats.shape, types.shape
```

```python
fig = plt.figure( figsize=[10,10])
plt.scatter( types[:,0], types[:,1])
plt.gca().set_aspect('equal')
```

```python
km = KMeans( n_clusters=5)
```

```python
km.fit( types)
```

```python
cluster_ids = km.predict( types)
```

```python
cluster_ids
```

```python
color_palettes = sns.color_palette( 'Set1',  5)
sns.palplot(color_palettes )
color_palettes = np.array( color_palettes)
```
