```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import time
from sklearn.preprocessing import OrdinalEncoder
import joblib as jb
```

```python
#df = pd.read_csv( "SALT_LAKE_CITY_-_POLICE_CASES_-2016.csv")
df = pd.read_csv( "SALT_LAKE_CITY_-_POLICE_CASES_2014.csv")
#df = pd.read_csv( "Salt_Lake_City_-_Police_Cases_2015.csv")
#df = pd.read_csv("Crimes_-_2018.csv")
```

```python
df.head(5)
```

```python
df = df[[ "UCR DESCRIPTION", 'OCC DATE', 'POLICE GRID', 'X COORDINATE', 'Y COORDINATE']].copy()
```

```python
len( df)
```

```python
df = df.dropna()
```

```python
len(df)
```

```python
df['UCR DESCRIPTION'].value_counts( )
```

```python
type_coults = df['UCR DESCRIPTION'].value_counts( )
```

```python
cum_sum = np.cumsum( type_coults.values)
```

```python
cum_density = cum_sum / np.sum( type_coults.values)
```

```python
MAJOR_PERCENTILE = 0.995
```

```python
major_type = type_coults.index [ :np.sum( cum_density < MAJOR_PERCENTILE) + 1].tolist()
```

```python
major_type
```

```python
grid_counts = df['POLICE GRID'].value_counts( )
```

```python
cum_sum = np.cumsum( grid_counts.values)
cum_density = cum_sum / np.sum( grid_counts.values)
```

```python
major_grid = grid_counts.index[ : np.sum( cum_density < MAJOR_PERCENTILE) + 1].tolist()
```

```python
major_grid
```

```python
df = df[ ( df['UCR DESCRIPTION'].map( lambda x : x in major_type) ) & ( df['POLICE GRID'].map( lambda x : x in major_grid))]
```

```python
len(df)
```

```python
df = df[ df['X COORDINATE'] > 100.0]
df = df[ df[ 'Y COORDINATE'] > 840000]
```

```python
df['OCCURRED_ts'] = df['OCC DATE'].map(  lambda x: time.mktime( time.strptime( x,'%m/%d/%Y %I:%M:%S %p' )))
```

```python
df = df.sort_values( 'OCCURRED_ts',axis=0)
```

```python
df.dtypes
```

```python
od_enc = OrdinalEncoder( dtype=np.int32)
```

```python
od_enc.fit( df['UCR DESCRIPTION'].values.reshape( -1,1))
```

```python
crime_types = od_enc.transform( df['UCR DESCRIPTION'].values.reshape( -1,1)).reshape(-1)
```

```python
df['crime_types'] = crime_types
```

```python
df['POLICE GRID'] = df['POLICE GRID'].astype( ('int32'))
```

```python
od_enc = OrdinalEncoder( dtype=np.int32)
grid_types = od_enc.fit_transform( df['POLICE GRID'].values.reshape( -1,1))
```

```python
df['grid_types'] = grid_types
```

```python
jb.dump( df,'SLC_CRIME.jlb')
```

```python
slc_ind = df[['grid_types', 'crime_types']].values
```

```python
slc_ind.shape
```

```python
slc_y = df['OCCURRED_ts'].values / ( 3600 * 24)
```

```python
slc_y = slc_y - np.min( slc_y)
```

```python
np.save('SLC_60k_inds', slc_ind)
np.save('SLC_60k_y_in_day', slc_y )
```

```python
event_tensor = {}
for i in range( len( slc_ind)):
    gId, cId = slc_ind[i]
    idx = (gId, cId)
    t = slc_y[i]
    if idx in event_tensor:
        event_tensor[idx].append( t)
    else:
        event_tensor[idx] = [t]

        
```

```python
seqs = list( event_tensor.values())
```
